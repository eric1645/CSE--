```python
import numpy as np
import matplotlib.pyplot as plt

#parameters
v = 1
n = 2
U_0 = 1
L = 10

#time nodes
t_max = (2*np.pi)/n
t_n = 512+1
t = np.linspace(0, t_max, t_n)
h = t[1] - t[0]
N = len(t)

#space nodes
y = np.linspace(0, L, 65)
hy = y[1] - y[0]
Ny = len(y)

#t=0  u
u = np.zeros(Ny).reshape(-1, 1)
u[-1] = U_0*np.cos(n*t[0])

#FTCS method
for i in range(1,N):
  u_y = np.zeros(Ny)
  u_y[0] = 0
  u_y[-1] = U_0*np.cos(n*t[i])
  for j in range(1,Ny-1):
    u_y[j] = u[j,-1] + v*h*(u[j+1, -1] -2*u[j, -1] +u[j-1,-1])/hy**2
  u_y_T = u_y.reshape(-1,1)
  u = np.hstack((u, u_y_T))

#
plt.figure(figsize=(12,15))

plt.subplot(3, 2, 1)
plt.plot(np.flip(u[:,0]),y)
plt.xlim(-1.1,1.1)
plt.xlabel("u")
plt.ylabel("y")
plt.title('2t=0')

plt.subplot(3, 2, 2)
plt.plot(np.flip(u[:,int((t_n-1)/4)]),y)
plt.xlim(-1.1,1.1)
plt.xlabel("u")
plt.ylabel("y")
plt.title('2t=pi/2')

plt.subplot(3, 2, 3)
plt.plot(np.flip(u[:,int((t_n-1)/2)]),y)
plt.xlim(-1.1,1.1)
plt.xlabel("u")
plt.ylabel("y")
plt.title('2t=pi')

plt.subplot(3, 2, 4)
plt.plot(np.flip(u[:,int(3*(t_n-1)/4)]),y)
plt.xlim(-1.1,1.1)
plt.xlabel("u")
plt.ylabel("y")
plt.title('2t=3pi/2')

plt.subplot(3, 2, 5)
plt.plot(np.flip(u[:,int(t_n-1)]),y)
plt.xlim(-1.1,1.1)
plt.xlabel("u")
plt.ylabel("y")
plt.title('2t=2pi')

plt.subplot(3, 2, 6)
plt.xlim(-1.1,1.1)
plt.plot(np.flip(u[:,0]),y, label = 'nt = 0')
plt.plot(np.flip(u[:,int((t_n-1)/4)]),y,label = 'nt = pi/2')
plt.plot(np.flip(u[:,int((t_n-1)/2)]),y, label = 'nt = pi')
plt.plot(np.flip(u[:,int(3*(t_n-1)/4)]),y, label = 'nt = 3pi/2')
plt.plot(np.flip(u[:,int((t_n-1))]),y, label = 'nt = 2pi')
plt.xlabel("u")
plt.ylabel("y")
plt.legend()
```

# Crank-Nicolson scheme

```python
import numpy as np
import matplotlib.pyplot as plt

#parameters
v = 1
n = 2
U_0 = 1
L = 10

#time nodes
t_max = (2*np.pi)/n
t_n = 512+1
t = np.linspace(0, t_max, t_n)
h = t[1] - t[0]
N = len(t)

#space nodes
y = np.linspace(0, L, 65)
hy = y[1] - y[0]
Ny = len(y)

a = v*h / (2*hy**2)

#B.C.
u = np.zeros((N, Ny))
u[:,0] = 0
u[0,-1] = U_0*np.cos(n*t[0])

# coefficient matrix
A = np.zeros((Ny-2,Ny-2))
B = np.zeros((Ny-2,Ny-2))

for i in range(Ny-2):
  A[i,i] = 2*a + 1
  B[i,i] = -2*a +1
  if i > 0:
    A[i, i-1] = -a
    B[i, i-1] = a
  if i < Ny-3:
    A[i, i+1] = -a
    B[i, i+1] = a

#Crank-Nicolson scheme
for i in range(N-1):
  u_col = u[i, 1:-1]
  b = B @ u_col
  b[0] += a * u[i, 0]
  b[-1] += a * u[i, -1]

  u_new = np.linalg.solve(A, b)
  u[i+1, 1:-1] = u_new
  u[i+1,-1] = U_0*np.cos(n*t[i+1])

#
plt.figure(figsize=(12,15))

plt.subplot(3, 2, 1)
plt.plot(np.flip(u[0,:]),y)
plt.xlim(-1.1,1.1)
plt.xlabel("u")
plt.ylabel("y")
plt.title('2t=0')

plt.subplot(3, 2, 2)
plt.plot(np.flip(u[int((t_n-1)/4),:]),y)
plt.xlim(-1.1,1.1)
plt.xlabel("u")
plt.ylabel("y")
plt.title('2t=pi/2')

plt.subplot(3, 2, 3)
plt.plot(np.flip(u[int((t_n-1)/2),:]),y)
plt.xlim(-1.1,1.1)
plt.xlabel("u")
plt.ylabel("y")
plt.title('2t=pi')

plt.subplot(3, 2, 4)
plt.plot(np.flip(u[int(3*(t_n-1)/4),:]),y)
plt.xlim(-1.1,1.1)
plt.xlabel("u")
plt.ylabel("y")
plt.title('2t=3pi/2')

plt.subplot(3, 2, 5)
plt.plot(np.flip(u[int(t_n-1),:]),y)
plt.xlim(-1.1,1.1)
plt.xlabel("u")
plt.ylabel("y")
plt.title('2t=2pi')

plt.subplot(3, 2, 6)
plt.xlim(-1.1,1.1)
plt.plot(np.flip(u[0,:]),y, label = 'nt = 0')
plt.plot(np.flip(u[int((t_n-1)/4),:]),y,label = 'nt = pi/2')
plt.plot(np.flip(u[int((t_n-1)/2),:]),y, label = 'nt = pi')
plt.plot(np.flip(u[int(3*(t_n-1)/4),:]),y, label = 'nt = 3pi/2')
plt.plot(np.flip(u[int((t_n-1)),:]),y, label = 'nt = 2pi')
plt.xlabel("u")
plt.ylabel("y")
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#exact solution
def exact_solution(y, t):
  eta = np.sqrt((n*y)/(2*v))
  return U_0*np.exp(-eta)*np.cos(n*t-eta)

#FTCS method
def FTCS(t, h, N, t_index):

  #t=0  u
  u = np.zeros(Ny).reshape(-1, 1)
  u[-1] = U_0*np.cos(n*t[0])

  for i in range(1,N):
    u_y = np.zeros(Ny)
    u_y[0] = 0
    u_y[-1] = U_0*np.cos(n*t[i])
    for j in range(1,Ny-1):
      u_y[j] = u[j,-1] + v*h*(u[j+1, -1] -2*u[j, -1] +u[j-1,-1])/hy**2
    u_y_T = u_y.reshape(-1,1)
    u = np.hstack((u, u_y_T))

  return u[:,t_index] # nt = pi/2

#Crank-Nicolson scheme
def CN(t, h, N, t_index):

  a = v*h / (2*hy**2)

  #B.C.
  u = np.zeros((N, Ny))
  u[:,0] = 0
  u[0,-1] = U_0*np.cos(n*t[0])

  # coefficient matrix
  A = np.zeros((Ny-2,Ny-2))
  B = np.zeros((Ny-2,Ny-2))

  for i in range(Ny-2):
    A[i,i] = 2*a + 1
    B[i,i] = -2*a +1
    if i > 0:
      A[i, i-1] = -a
      B[i, i-1] = a
    if i < Ny-3:
     A[i, i+1] = -a
     B[i, i+1] = a

  for i in range(N-1):
    u_col = u[i, 1:-1]
    b = B @ u_col
    b[0] += a * u[i, 0]
    b[-1] += a * u[i, -1]

    u_new = np.linalg.solve(A, b)
    u[i+1, 1:-1] = u_new
    u[i+1,-1] = U_0*np.cos(n*t[i+1])

  return u[t_index,:] # nt = pi/2

def error(exact, numerical):
  return np.sqrt(np.mean((exact-numerical)**2))

#parameters
v = 1
n = 2
U_0 = 1
L = 10

#time nodes
t_max = (2*np.pi)/n
t_n = [5000, 5500, 6000]

#space nodes
y = np.linspace(0, L, 129)
hy = y[1] - y[0]
Ny = len(y)

error_FTCS= []
error_CN = []
delta_t = []

for i in t_n:
  t = np.linspace(0, t_max, i)
  h = t[1] - t[0]
  N = len(t)

  t_index = np.argmin(np.abs(n*t - (np.pi)))

  result_FTCS = FTCS(t,h,N,t_index)
  result_CN = CN(t,h,N,t_index)
  exact = exact_solution(np.flip(y),t[t_index])

  delta_t.append(h)
  error_FTCS.append(error(exact,result_FTCS))
  error_CN.append(error(exact,result_CN))

plt.loglog(delta_t, error_FTCS, label='FTCS')
plt.loglog(delta_t, error_CN, label='CN')
plt.legend()

slope_ftcs = np.polyfit(np.log(delta_t), np.log(error_FTCS), 1)[0]
slope_cn = np.polyfit(np.log(delta_t), np.log(error_CN), 1)[0]

print(f"FTCS   (): {slope_ftcs:.3f}")
print(f"Crank-Nicolson   (): {slope_cn:.3f}")
```
